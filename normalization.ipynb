```python
import pandas as pd
from sklearn import preprocessing

x = df[['pl_orbper','st_dist','st_optmag','st_mass']].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_temp = pd.DataFrame(x_scaled, columns=['pl_orbper','st_dist','st_optmag','st_mass'], index = df.index)
df[['pl_orbper','st_dist','st_optmag','st_mass']] = df_temp
```
