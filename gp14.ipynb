```python
#HYPOTHESIS TESTING
```

```python
from statsmodels.stats import weightstats as po
```

```python
ztest,pvals=po.ztest(df['Rating'].sample(n=100,random_state=1),value=4,alternative="larger")#value is of the null hypothesis; larger represents alternate hypothesis.
print("p-value",pvals)
```

```python
if pvals<=0.05:
    print("We reject NULL hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

```python
#As p value is very much less than 0.05,we can conclude that the Null hypothesis can be Rejected. 
```
